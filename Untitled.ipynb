```python
import numpy as np
import matplotlib.pyplot as plt
from utils.metrics import *

tval = np.array( [1,2,3,4,5])
est = np.array([2,3,4,5,6])
est2 = np.array([2.6,2.8,3,3.2,3.4])

```

```python
input_val = np.array( [[t] for t in range(5)])
print("MAE OF TVAL OVER EST", MAE(tval, est))
print("MAE OF TVAL OVER EST2", MAE(tval, est2))
print("MSE OF TVAL OVER EST", MSE(tval, est))
print("MSE OF TVAL OVER EST2", MSE(tval, est2))
print("STD_RATIO OF TVAL OVER EST", 1/2*(np.std(tval)/np.std(est) + np.std(est)/np.std(tval)))
print("STD_RATIO OF TVAL OVER EST2", 1/2*(np.std(tval)/np.std(est2) + np.std(est2)/np.std(tval)))
print("SLOPE COMPARISON OF TVAL OVER EST", 5*(get_slope(input_val, tval) -get_slope(input_val, est )))
print("SLOPE COMPARISON OF TVAL OVER EST2", 5*(get_slope(input_val, tval) -get_slope(input_val, est2 )))
```

```python
# graph
x1 = [k+1 for k in range(5)]
X = np.array([[k] for k in range(5)])

plt.plot(x1, tval, 'k', label='true')
plt.plot(x1, est,'r--', label='est-1')
plt.plot(x1, est2,'b--', label='est-2')
plt.title('Similar MAE/MSE but different Trend')
plt.xticks(range(1,6))
plt.xlabel('steps')
plt.ylabel('value')
plt.legend()
plt.savefig('./graph.png')
plt.show()
```
